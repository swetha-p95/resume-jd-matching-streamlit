```python
#resume custom ner training

import spacy
import pickle
import random
import json
from spacy.util import minibatch, compounding
from spacy.training import Example
from spacy import displacy


#training_data = pickle.load(open("train_data.pkl","rb"))
nlp = spacy.blank("en")

with open("C:\\Users\\HP\\Desktop\\DUK\\S2\\NLP\\project_final\\annotations_resume_110.json", "r") as file:
    data = json.load(file)

# Convert JSON data to spaCy training format
training_data = []
#print(data)
#print(data[0][1])
for item in data:
    training_data.append((item[0],  item[1]))

examples = [Example.from_dict(nlp.make_doc(text), annotations) for text, annotations in training_data]



def train_model(training_data):
    if 'ner' not in nlp.pipe_names:
      nlp.create_pipe("ner")
      ner = nlp.add_pipe("ner",last=True)
    else:
      ner = nlp.get_pipe('ner')
    count = 0
    for text, annotations in training_data:
      for ent in annotations["entities"]:
         ner.add_label(ent[2])
      count +=1
      #print(annotations)
    # print(count) 
     
    other_pipes = [pipe for pipe in nlp.pipe_names if pipe != "ner"]
    with nlp.disable_pipes(*other_pipes):
        optimizer = nlp.initialize()
        for itn in range(500):
            #print("Starting iteration",itn)
            random.shuffle(training_data)


            losses = {}
            try:
              for batch in minibatch(examples, size=compounding(4.0, 32.0, 1.001)): #4,32
                 nlp.update(batch, sgd=optimizer, drop=0.35, losses=losses)
            except ValueError:
                 pass
        print("Resume model training completed!")


train_model(training_data)
# print("\n\n")
# print(nlp.get_pipe("ner").labels)
nlp.to_disk("new_model_resume")
```
